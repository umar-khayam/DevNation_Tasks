```python
import pandas as pnd
import operator
```

```python
df=pnd.read_csv('olympics.csv',index_col=0,skiprows=1)
df.head()
```

```python
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]},inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]},inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]},inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]},inplace=True)
names_ids=df.index.str.split('\s\(')
df.index=names_ids.str[0]
df['ID']=names_ids.str[1].str[:3]
df=df.drop('Totals')
df.head()
```

```python
def name_of_country():
    a=df[df['Gold']==df['Gold'].max()].index.values
    return a[0]
name_of_country()
```

# Question 2

```python
def medal_count_difference(): 
    new_=df[['Gold','Gold.1']]
    new_['Sub']=new_['Gold']-new_['Gold.1']
    new_['Total']=new_['Gold']+new_['Gold.1']
    new_['Final']=new_['Sub']/new_['Total']
    a=new_[new_['Final']==new_['Final'].max()].index.values
    return a[0]
medal_count_difference()
```

```python
def difference():   
    new_df=df[operator.and_(df['Gold']>=1, df['Gold.1'])]
    new_df=new_df[['Gold','Gold.1']]
    new_df['Sub']=new_df['Gold']-new_df['Gold.1']
    new_df['Total']=new_df['Gold']+new_df['Gold.1']
    new_df['Final']=new_df['Sub']/new_df['Total']
    a=new_df[new_df['Final']==new_df['Final'].max()].index.values
    return a[0]
difference()
```

```python
def series_of_data():
    d=df.iloc[:,11:14]
    a=[]
    for i in range(len(d)):
        a.append(d['Gold.2'][i]*3+d['Silver.2'][i]*2+d['Bronze.2'][i])
    a=pnd.Series(a,index=df.index)
    return a
series_of_data()
```

```python
def population_cty():
    population_data=data.iloc[:,8:15]
    data['Max']=population_data.max(axis=1)
    data['Min']=population_data.min(axis=1)
    data['Change']=data['Max']-data['Min']
    a=list(data[data['Change']==data['Change'].max()]['CTYNAME'])
    return a[0]
population_cty()
```

```python
def stname():
    new_data=data[data['CTYNAME'].str.contains('Washington')]
    new_data=new_data[operator.or_(new_data['REGION']==1 , new_data['REGION']==2)]
    new_data=new_data[new_data['POPESTIMATE2015']>new_data['POPESTIMATE2014']]
    new_data.sort_index(axis=1)
    return new_data[['STNAME','CTYNAME']]
stname()
```
