```python
import pandas as pnd
```

```python
df=pnd.read_csv('olympics.csv',index_col=0,skiprows=1)
df
```

```python
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]},inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]},inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]},inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]},inplace=True)
names_ids=df.index.str.split('\s\(')
df.index=names_ids.str[0]
df['ID']=names_ids.str[1].str[:3]
df=df.drop('Totals')
df.head()
```

```python
def name_of_country():
    a=df[df['Gold']==df['Gold'].max()].index.values
    return a[0]
name_of_country()
```

```python
def series_of_data():
    d=df.iloc[:,11:14]
    a=[]
    for i in range(len(d)):
        a.append(d['Gold.2'][i]*3+d['Silver.2'][i]*2+d['Bronze.2'][i])
    a=pnd.Series(a,index=df.index)
    return a
series_of_data()
```
