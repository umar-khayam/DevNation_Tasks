```python
import pandas as pnd
import numpy as np
```

# Question 01

```python
energy=pnd.read_excel('Energy+Indicators.xls',header=16)
energy=energy.iloc[1:,2:]
energy=energy.replace('...',np.nan)
energy['Energy Supply']=energy['Energy Supply']*1000000
energy.rename(columns={'Unnamed: 2':'Country','Renewable Electricity Production':'% Renewable' },inplace=True)
energy['Country'] = energy['Country'].str.extract('(^[a-zA-Z\s]+)', expand=False).str.strip()
energy.replace({"Republic of Korea": "South Korea","United States of America": "United States",
              "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
              "China, Hong Kong Special AdministrativeRegion": "Hong Kong"},inplace=True)
```

```python
gdp=pnd.read_csv('world_bank.csv',header=None)
gdp.rename(columns=gdp.iloc[4],inplace=True)
gdp=gdp.iloc[5:]
gdp=gdp.reset_index()
gdp.rename(columns={'Country Name': 'Country'}, inplace=True)
gdp.replace({"Korea, Rep.": "South Korea", "Iran, Islamic Rep.": "Iran", "Hong Kong SAR, China": "Hong Kong"},inplace=True)
```

```python
ScimEn= pnd.read_excel('scimagojr-3.xlsx')
```

```python
def data_merging():
    new_dataset= pnd.merge(pnd.merge(energy, gdp,on="Country" ),ScimEn ,on="Country",how='inner')
    new_dataset.set_index("Country",inplace = True)
    new_dataset=new_dataset.sort_values('Rank')
    new_dataset=new_dataset[['Rank', 'Documents', 'Citable documents', 
                             'Citations', 'Self-citations', 'Citations per document',
                             'H index', 'Energy Supply','Energy Supply per capita','% Renewable', 2006.0, 
                             2007.0, 2008.0, 2009.0, 2010.0, 2011.0, 2012.0, 2013.0, 2014.0, 2015.0]]
    new_dataset=new_dataset.drop_duplicates(subset=['Rank'])
    return new_dataset[new_dataset['Rank']<=15]
data_merging()
```

# Question 02

```python
def Question_02():
    return len(new_dataset)-15
Question_02()
```

# Question 03

```python
def Question_03():
    gdp['avgGDP'] = gdp[[2006.0, 2007.0, 2008.0, 2009.0, 2010.0, 2011.0, 2012.0, 2013.0, 2014.0, 2015.0]].mean(axis=1)
    series_avg=gdp.sort_values("avgGDP",ascending=False)["avgGDP"]
    return series_avg[:15]
Question_03()
```

# Question 04

```python
gdp=gdp.fillna(0)
def largest_6th():
    gdp['avgGDP']=gdp['avgGDP'].astype('int64')
    data_set=gdp.nlargest(6,'avgGDP')
    data_set['Difference']=data_set[2015]-data_set[2006]
    return data_set.iloc[-1,-1]
largest_6th()
```

# Question 05

```python
def mean_energy_supply():
    energy['Energy Supply per capita']=energy['Energy Supply per capita'].astype('float64')
    return energy['Energy Supply per capita'].mean()
mean_energy_supply()
```

# Question 06

```python

def country_max_Renewable():
    tupple=energy.loc[energy['% Renewable'].idxmax()]
    return (tupple['Country'],tupple['% Renewable'])
country_max_Renewable()
```

# Question 07

```python
def country_max_citation_ration():
    ScimEn['Ratio citations']=(ScimEn['Self-citations']/ScimEn['Citations'])*100
    tupple=ScimEn.loc[ScimEn['Ratio citations'].idxmax()]
    return (tupple['Country'],tupple['Ratio citations'])
country_max_citation_ration()
```

```python
def populas_country():
    energy['Populas']=energy['Energy Supply']/energy['Energy Supply per capita']
    third_most=energy.nlargest(3,'Populas')
    return third_most.iloc[-1]['Country']
populas_country()
```
